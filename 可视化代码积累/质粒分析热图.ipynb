```python
from collections import defaultdict

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.colors import ListedColormap
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.colors import to_rgb


import os
```

```python

base_path = '/Users/yao/Nutstore Files/Mac2PC/Huang-Bacteria/最终结果整理/co-occurrence'

fig_save_path = f'{base_path}/heat-map-dir'

if not os.path.exists(fig_save_path):
    os.mkdir(fig_save_path)
```

```python
def read_res_data(res_data: str):

    gene_list = []

    year_gene_cor_list = []
    
    num = 0
    norm_len = 0
    with open(res_data) as f:
        #f.readline()
        year_list = f.readline().strip().split('\t')[1:]
        for line in f:
            
            l = line.strip().split('\t')
            
            gene = l[0]
            
            if not gene:
                continue

            gene_list.append(gene)
            
            value_list = []
            for i in l[1:]:
                value_list.append(float(i))
            
            if value_list:
                year_gene_cor_list.append(value_list)

    print(f'Gene Count: {len(gene_list):,}')
    return year_list, gene_list, year_gene_cor_list
```

```python
# figure test 


# read file
pla_res_file = f'{base_path}/plasmid-ARG.hotmap.txt'

pla_res_fig = f'{fig_save_path}/plasmid-ARG.hotmap.png'

pla_gene_list, res_gene_list, pla_res_value_list = read_res_data(pla_res_file)



# draw plot

#colors = [(69,117,183,), (212,235,241), (247,251,200),(253,210,141),(215,46,45)]
colors = [(15,24,245), (250,250,250), (255,33,18)]


# 将颜色列表转换为Matplotlib中的颜色格式
colors = [(r/255., g/255., b/255.) for r, g, b in colors]

# 创建自定义colormap并反转
# .reversed()
cmap = LinearSegmentedColormap.from_list('my_colormap', colors, N=256)


font_size = 24


year_gene_cor_matrix = np.array(pla_res_value_list)

# 换最小值
year_gene_cor_matrix[year_gene_cor_matrix==-0.579464262] = -0.846728105


plt.subplots(figsize=(38, 28))
sns.set_theme()

ax = sns.heatmap(year_gene_cor_matrix, cbar=True, cmap=cmap,
                     cbar_kws={'location': 'top', 'shrink': 0.3})

plt.xticks([i+0.5 for i in range(len(pla_gene_list))], pla_gene_list, rotation=270, size=font_size, fontfamily='Times New Roman')
plt.yticks([i+0.5 for i in range(len(res_gene_list))], res_gene_list, rotation=0, size=font_size, fontfamily='Times New Roman')

colorbar = ax.collections[0].colorbar
colorbar.outline.set_color('black')
colorbar.outline.set_linewidth(1)
colorbar.ax.tick_params()

# set the ticks for colorbar
#colorbar.set_ticks([0, 0.5, 1])
#colorbar.set_ticklabels(['0', '0.5', '1'])

plt.savefig(pla_res_fig, dpi=360, bbox_inches='tight')

print(f'{pla_res_fig} saved.')
```
