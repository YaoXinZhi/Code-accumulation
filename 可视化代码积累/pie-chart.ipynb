```python
import matplotlib.pyplot as plt

# Create a sample data set
labels = ['Class A', 'Class B', 'Class C', 'Class D']
sizes = [20, 30, 25, 25]

# Define colors for each class
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Plot the pie chart
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%',
        startangle=90, textprops={'fontsize': 14})
ax1.axis('equal')

# Add title and legend
plt.title('Sample Pie Chart', fontsize=16)
plt.legend(labels, loc="best", fontsize=12)

# Show the plot
plt.show()
```

```python
import os
from collections import defaultdict

import matplotlib.pyplot as plt
import numpy as np
```

```python
base_path = '/Users/yao/Nutstore Files/Mac2PC/Huang-Bacteria/世界地图+饼图'

data_file = f'{base_path}/世界地图-data.txt'

fig_save_path = f'{base_path}/pie-figs'
```

```python
if not os.path.exists(fig_save_path):
    os.mkdir(fig_save_path)
```

```python

location_to_species_count = defaultdict(dict)
with open(data_file) as f:
    head_line = f.readline().strip().split('\t')
    species_list = head_line[4:]
    print(species_list)
    for line in f:
        l = line.strip().split('\t')
        
        country = l[0]

        species_count = l[4:]
        
        for idx, count in enumerate(species_count):
            if count:
                species = species_list[idx]
                
                #print(f'{species_list[idx]}-{count}')
                location_to_species_count[country][species] = count
        #print(location_to_species_count)
        #input()
```

```python
species_to_color = {
    'homo': '#2F7FC1',
    'Swine': '#D8383A',
    'chicken': '#C497B2',
    'cow': '#FFBE7A',
    'dog': '#96C37D',
    'cat': '#F3D266',
    'blowfly': '#999999',
    'enviroment': '#A9B8C6',
    'not collection': '#F8F3F9',
}
```

```python
def pie_chart(location: str, species_count: dict, 
              species_to_color: dict, save_file: str):
    
    num_list = []
    color_list = []
    for species, count in species_count.items():
        color = species_to_color[species]
        
        num_list.append(int(count))
        color_list.append(color)
        
    num_list = np.array(num_list)

    
    plt.pie(num_list,
            colors=color_list,
            #startangle=90, #pctdistance=0.85,
            #wedgeprops={'linewidth': 0.6, 'edgecolor': 'black'}
            #                        'alpha': 0.8}
           )
    
    plt.title(location, fontsize=16)
    #plt.show()
    
    plt.savefig(save_file)
    print(f'{save_file} saved.')
```

```python

for location, species_count in location_to_species_count.items():
    
    save_file = f'{fig_save_path}/{location}-pie.png'
    
    #save_file = f'/Users/yao/Desktop/pie-figs/{location}-pie.png'
    
    pie_chart(location, species_count, species_to_color, save_file)

```
