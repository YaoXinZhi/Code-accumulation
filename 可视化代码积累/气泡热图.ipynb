```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
```

```python

N = 10
M = 11
ylabels = ["".join(np.random.choice(list("PQRSTUVXYZ"), size=7)) for _ in range(N)]
xlabels = ["".join(np.random.choice(list("ABCDE"), size=3)) for _ in range(M)]

x, y = np.meshgrid(np.arange(M), np.arange(N))
s = np.random.randint(0, 180, size=(N,M))
c = np.random.rand(N, M)-0.5

# statrt to plot
fig, ax = plt.subplots()

R = s/s.max()/2
circles = [plt.Circle((j,i), radius=r) for r, j, i in zip(R.flat, x.flat, y.flat)]
col = PatchCollection(circles, array=c.flatten(), cmap="RdYlGn")
ax.add_collection(col)

ax.set(xticks=np.arange(M), yticks=np.arange(N),
       xticklabels=xlabels, yticklabels=ylabels)
ax.set_xticks(np.arange(M+1)-0.5, minor=True)
ax.set_yticks(np.arange(N+1)-0.5, minor=True)
ax.grid(which='minor')


fig.colorbar(col)
plt.show()
```

```python
"""

这是一个使用Matplotlib绘制气泡图的示例代码。它使用了numpy.meshgrid()方法生成网格点坐标，然后使用matplotlib.pyplot.Circle()方法绘制圆形，并用颜色表示数据。

下面是一些对代码的解释：

N和M分别表示行数和列数。
ylabels和xlabels分别表示行标签和列标签，它们是由7个字母随机组合而成的字符串。

x和y是由一维数组生成的网格点坐标，np.meshgrid()方法可以将两个一维数组转换为二维数组。
s是一个随机矩阵，它决定了每个圆圈的大小（半径）。
c是一个随机矩阵，它决定了每个圆圈的颜色。


plt.Circle()方法用于绘制每个圆圈，其中j和i表示圆心坐标，r表示半径。
PatchCollection()方法将所有圆圈添加到集合中，并根据颜色矩阵c进行着色。
fig.colorbar()方法在图像旁边添加了一个颜色条，该颜色条展示了颜色映射。
运行上述代码后，就可以看到一个带有随机颜色和半径的气泡图，其中每个圆圈的大小和颜色都代表了数据。

"""
```

```python
def read_matrix_file(matrix_file: str):
    
    # size, abs of value
    s_matrix = []
    # color, value
    c_matrix = []
    
    y_labels = []
    with open(matrix_file) as f:
        x_labels = f.readline().strip().split('\t')[1:]
        for line in f:
            l = line.strip().split('\t')
            y_labels.append(l[0])
            
            value_list = [float(value) for value in l[1:]]
            
            s_matrix.append(list(map(abs, value_list)))
            c_matrix.append(value_list)
    
    s_matrix = np.array(s_matrix)
    c_matrix = np.array(c_matrix)
    
    
    N,M = s_matrix.shape
    x, y = np.meshgrid(np.arange(M), np.arange(N))
    
    
    return N, M, y_labels, x_labels, x, y, s_matrix, c_matrix
```

```python
# args file
input_file = f'{base_path}/plasmid-ARG.hotmap.txt'

save_file = f'{base_path}/plasmid-ARG.bubble-heat.png'

# DATA STATISTICS

fontsize = 24

N, M, ylabels, xlabels, x, y, s, c = read_matrix_file(args_file)

# statrt to plot
fig, ax = plt.subplots(figsize=(58, 24))

#R = s/s.max()/2
R = s/s.max()/1.7

circles = [plt.Circle((j,i), radius=r) for r, j, i in zip(R.flat, x.flat, y.flat)]
col = PatchCollection(circles, array=c.flatten(), cmap="RdYlGn")
ax.add_collection(col)

ax.set(xticks=np.arange(M), yticks=np.arange(N),
       xticklabels=xlabels, yticklabels=ylabels)
ax.set_xticks(np.arange(M+1)-0.5, minor=True)
ax.set_yticks(np.arange(N+1)-0.5, minor=True)
ax.grid(which='minor')


plt.xticks(rotation=270, fontsize=fontsize, fontfamily='Times New Roman')
plt.yticks(fontsize=fontsize, fontfamily='Times New Roman')

fig.colorbar(col)

# 添加颜色条，并设置其大小及刻度及字体属性
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=fontsize, direction='inout', length=10, width=1, pad=8)
#cbar.set_label('Intensity', fontsize=fontsize, fontfamily='Times New Roman')


plt.savefig(save_file, dpi=420, bbox_inches='tight')

print(f'{save_file} saved.')
#plt.show()
```
